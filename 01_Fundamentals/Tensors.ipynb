# Creating tensors

```python
# Zeros
a=torch.zeros((2,3))
a
```

```python
# Ones 
b=torch.ones(((2,3)))
b
```

```python
# random numbers between 0 to 1

a= torch.rand((2,3))         #uniform [0,1]
a
```

```python
# random numbers (normal distribution)
b=torch.randn((2,3))
b
```

```python
# a x b matrix of a specific number
a=torch.full((2,3),7)
a
```

```python
# identity matrix
a= torch.eye(3)
a
```

```python
# integers from x to y-1
a=torch.arange(0,6)
a
```
