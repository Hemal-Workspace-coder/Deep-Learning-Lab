# Read an Image

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Read the image
img = mpimg.imread("puppy.jpg")

# Display the image
plt.imshow(img)
plt.axis("off")  # Hide axes
plt.show()
```

```python
img.shape
```

1536 → Height of the image in pixels (number of rows).

2048 → Width of the image in pixels (number of columns).

3 → Number of channels per pixel.

Typically this means RGB (Red, Green, Blue).

Each pixel has 3 values, one for each color channel.

```python
# Converting an img into numpy array
print(img)
```

```python
# Displaying the image from numpy array

img_plot=plt.imshow(img)
plt.show()
```

# Resize an image using pillow library

```python
from PIL import Image
```

```python
img=Image.open('puppy.jpg')
img_resized = img.resize((200,200))
```

```python
img_resized.save('pyppy_resized.jpg')
```

```python
# Displaying the resized image intp a numpy format
img_res = mpimg.imread("pyppy_resized.jpg")
plt.imshow(img_res)
plt.axis("off")  # Hide axes
plt.show()
```

```python
# Shape ensure
print(img_res.shape)
```

# Converting RGB images to GrayScale image using OpenCV

```python
import cv2
```

```python
img=cv2.imread("puppy.jpg")
```

```python
img.shape
```

```python
grayscale_img=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```

```python
plt.imshow(grayscale_img, cmap='gray')
plt.axis("off")   # Hide axes
plt.show()
```
