```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

# Defininf the Input Text and Prepare Charater Set

```python
text= "This is GeeksforGeeks a software training institute"

# 1. set(text) extracts unique characters (e.g., duplicates like 'e' and 's' are removed)
# 2. list(...) converts the set back into a list format so it can be ordered
# 3. sorted(...) sorts the unique characters alphabetically (including spaces and capitals)

chars= sorted(list(set(text)))

# Create a dictionary mapping each unique character to a unique sequential integer ID

char_to_index = {char: i for i, char in enumerate(chars)}

# Create the inverse dictionary mapping each integer ID back to its corresponding character
index_to_char={i: char for i, char in enumerate (chars)}
```

# Creating Sequences and Labels

```python
seq_length = 3  # Sliding window size (step_size)
sequences = []  # Hold training input samples
labels = []     # Hold the correct expected outputs for those inputs

for i in range(len(text) - seq_length):
    seq = text[i:i + seq_length]  # string slicing
    label = text[i + seq_length]  # Grab the single char after the seq char
    sequences.append([char_to_index[char] for char in seq])
    labels.append(char_to_index[label]) # Learns the letter that come after each slice 

X = np.array(sequences)
y = np.array(labels)
```

# Creating sequences and labels

```python
# Converting X and y into one-hot encoded tensors
x_one_hot= tf.one_hot(X, len(chars))
y_one_hot= tf.one_hot(y, len(chars))
```

# Building the RNN model

```python
# Simple RNN

model= Sequential()
model.add(SimpleRNN(50, input_shape=(seq_length, len(chars)), activation='relu'))
model.add(Dense(len(chars), activation='softmax'))
```

# Compiling and Training the Model

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x_one_hot, y_one_hot, epochs=100)
```

# Generating New Text using the trained model

```python
start_seq = "This is G"
generated_text = start_seq

for i in range(50):
    x = np.array([[char_to_index[char] for char in generated_text[-seq_length:]]])
    x_one_hot = tf.one_hot(x, len(chars))
    prediction = model.predict(x_one_hot)
    next_index = np.argmax(prediction)
    next_char = index_to_char[next_index]
    generated_text += next_char

print("Generated Text:")
print(generated_text)
```
